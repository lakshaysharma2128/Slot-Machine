```python
import random #This "random" module will generate the numbers in a random order

MAX_LINES = 3 #Global constant that's not gonna change 
MAX_BET = 1000 
MIN_BET = 10

ROWS = 3
COLS = 3 #columns 

symbols_count = { #A dictionary for the symbols
    "A":12,   # Here the numbers tells us bout the count of symbols in each row
    "B":15,
    "C":18,
    "D":21
}

symbol_value = {
    "A":12,  #This will multiply the winnings of the user
    "B":9,
    "C":6,
    "D":3  
}

def check_winnings(columns, lines, bet, values):
    winnings = 0
    winning_lines = []
    for line in range(lines):
        symbol = columns[0][line]
        for column in columns:
            symbol_to_check = column[line]
            if symbol == symbol_to_check:
                break
        else:
            winnings += values[symbol]*bet
            winning_lines.append(line + 1)
    
    return winnings, winning_lines
    
def get_slot_machine_spin(rows , cols , symbols): #This is gonna be the generator
    all_symbols = [] #This is gonna be our list
    for symbol , symbol_count in symbols.items(): #Using ".items" gives you both the key and the value associated with the dictionary 
        for _ in range(symbol_count): # "_" is used when no variable is required or we just want to loop it but not a specific numbers of time 
            all_symbols.append(symbol)   

    columns = [] #A nested list would be formed that'll contains all the elements (rows , columns , symbols)
    for _ in range(cols):
            #current_pool = base_pool[:] 
            current_symbols = all_symbols[:] # [:] this works as a copy and separates the main value from the copy one the operator refers an Slice operator
            column = []
            for _ in range(rows):
                value = random.choice(current_symbols)
                #current_pool.remove(value) 
                current_symbols.remove(value) #.remove is used to remove an element from the list 
                column.append(value) #We use append to add a variable in a list  
            
            columns.append(column) 
        
    return columns
                
def print_slot_machine(columns):
       for row in range(len(columns[0])):
           for i, column in enumerate(columns): # enumerate():- keep track of both the item and its index (position) at the same time.
               if i !=  len(columns) -1: # This the max index we have 
                   print(column[row], end = " | " ) #We use pipe operators or "|" to    
               else:
                   print(column[row] , end = "" ) #end operator tells the print statement what to end the line with
           
           print() #That extra "print()" is needed to move to the next line after printing one row of the slot machine.
            #Intend this print statement with for loop

def deposit(): #responsible for users input to deposit the amount 
    while True: # Continuously asking the user to enter a valid amount until we get one
        amount = input("What would you like to deposit? $")
        if amount.isdigit():#isdigit is gonna tell us whether the input is a whole number or not
            amount = int(amount)
            if amount >= 10:
                break 
            else:
                print("Amount must be greater that 10$.")
        else:
             print("Please enter a number.")
    return amount

def get_number_of_lines(): #The user will choose from 1 - 3 lines since we have 3 lines 
       while True: # Continuously asking the user to choose a line
        lines = input("Enter the number of lines to bet on (1-" + str(MAX_LINES) + ")?")
        if lines.isdigit():#isdigit is gonna tell us which line is selected 
            lines = int(lines)
            if 1 <= lines <= MAX_LINES:
                break 
            else:
                print("Enter a valid number of lines")
        else:
             print("Please enter a line")
       return lines
    
def get_bet(): #The place will be bet between MIN_BET and MAX_BET
       while True: # Asks the user to place a bet 
        amount = input("What would you like to bet on each line")
        if amount.isdigit():#isdigit is gonna tell us bout placing a bet  
            amount = int(amount)
            if MIN_BET <= amount <= MAX_BET:
                break 
            else:
                print(f"Amount must be between ${MIN_BET} - ${MAX_BET}.")#putting "f" before the string and it will convert a variable into string if it can't be converted [3.6 or higher versions only] 
        else:
             print("Please enter a bet")
       return amount

def spin(balance):
    lines = get_number_of_lines()
    while True:
        bet = get_bet()
        total_bet = bet * lines

        if total_bet > balance: 
            print(f"You do not have enough balance to bet, your current balance is ${balance}")
        else:
            break
        
    print(f"You are betting ${bet} on {lines} lines .\n Your total bet will be: ${total_bet}")
    
    slots = get_slot_machine_spin(ROWS, COLS, symbols_count) 
    print_slot_machine(slots)
    winnings, winning_lines = check_winnings(slots, lines, bet, symbol_value)
    print(f"You won ${winnings}.")
    print(f"You won on lines: ", *winning_lines) # "*" asterisk operators are used to unpack a list/tuple
                                                 # *winning_lines will tell you at what lines will you win 
    return winnings - total_bet

def main(): #For re-running the main function 
    balance = deposit()
    while True:
        print(f"Current balance is: ${balance}")
        answer = input("Wanna continue \n Yes or No: ")
        if answer == "No":
            break 
        balance += spin(balance)

    print(f"You left with ${balance}")
            
main()  
```
